```python
import numpy as np
import numpy.linalg as npla
```

## QR Factorization Example

### So much easier than doing the Gram-Schmidt Method...!

***Requirement:*** *Any real square matrix A can be factored this way*

```python
A = np.array([[1,2,4],[0,0,5],[0,3,6]])

###########################
# Try these different examples also:
A = np.array([[1,0],[1,4]])
A = np.array([[-1,22,47],[9,-4,154],[0.4,3.3,-6]])
A = np.round(10*np.random.rand(5000,5000))

Q, R = npla.qr(A)

# To set precision of prited outputs:
np.set_printoptions(precision=3)

print("A:\n", A)
print("\nQ:\n", Q,"\nR:\n", R, "\n\nQ@R (compare with matrix A):\n", Q@R , '\n')

print("\nRelative residual norm:", npla.norm(A - Q@R)/npla.norm(A))
```

## Test for Orthogonality of a Matrix

**Arguments:**

**M**: Input matrix

**tolerance**: (optional) Set tolerance of comparison (ideally, this is zero, but we'll default it to $10^{-8}$)

```python
def ortho(M, tolerance = 1e-8):
    # First, check if M is a square matrix!
    m,n = M.shape
    assert m==n, 'The matrix is not square'
    
    # Check if M is orthogonal by doing this test!
    if np.all( np.abs(M@M.T - np.eye(n)) <= tolerance ):
        print("It's Orthogonal!")
    else:
        print("It's NOT Orthogonal!")
```
