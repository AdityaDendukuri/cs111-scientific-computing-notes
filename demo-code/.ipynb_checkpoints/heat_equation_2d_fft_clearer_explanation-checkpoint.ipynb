# CS111 Project Template — 2D Heat Equation with the Fourier Basis and FFT

We solve the periodic 2D heat equation

$$u_t = \kappa (u_{xx} + u_{yy}), \qquad (x,y) \in [0,2\pi]^2$$

using the **Fourier basis** and the **FFT**.

This notebook is written for students who know the basics of finite differences, basis functions, and linear algebra, but are seeing Fourier methods for the first time.


## 1. Why are we changing basis?

In linear algebra, a basis is a set of building blocks used to represent vectors.

For example, in the plane we usually use the standard basis

$$e_1 = \begin{bmatrix}1\\0\end{bmatrix}, \qquad e_2 = \begin{bmatrix}0\\1\end{bmatrix}.$$

Then any vector can be written as

$$v = c_1 e_1 + c_2 e_2.$$

The numbers $c_1,c_2$ are the coordinates of the vector **in that basis**.

The same idea works for functions. A function can be represented using different basis functions.


## 2. Our current viewpoint: sampled values on a grid

In finite differences, we represent a function by its values at grid points.

So on a grid $x_0,x_1,\dots,x_{n-1}$, our data looks like

$$u = [u(x_0), u(x_1), \dots, u(x_{n-1})].$$

This is the viewpoint we already know.

But sometimes another basis makes the differential operator much simpler. That is exactly what happens for the heat equation with periodic boundary conditions.


## 3. The Fourier basis

For periodic functions, a natural basis is the Fourier basis.

In 1D, the basic Fourier modes are

$$e^{ikx}, \qquad k \in \mathbb{Z}.$$

You can also think of these as being built from sines and cosines.

In 2D, the Fourier basis functions are

$$e^{i(kx+\ell y)}, \qquad k,\ell \in \mathbb{Z}.$$

So instead of representing a function by point values, we can represent it by Fourier coefficients:

$$u(x,y) = \sum_{k,\ell} \hat u_{k,\ell} e^{i(kx+\ell y)}.$$

The numbers $\hat u_{k,\ell}$ tell us how much of each Fourier mode is present.


## 4. Why the Fourier basis is useful for the heat equation

The heat equation is

$$u_t = \kappa (u_{xx}+u_{yy}).$$

The key fact is that Fourier modes are eigenfunctions of the Laplacian:

$$\Delta e^{i(kx+\ell y)} = -(k^2+\ell^2)e^{i(kx+\ell y)}.$$

This is exactly what we want.

In physical space, the Laplacian couples nearby grid points together.

In the Fourier basis, the Laplacian just multiplies each mode by the scalar $-(k^2+\ell^2)$.

So the PDE becomes much simpler after a change of basis.


## 5. What the FFT does

The **Fast Fourier Transform (FFT)** is an algorithm that computes Fourier coefficients efficiently.

You do **not** need to implement FFT from scratch for this project.

You only need to know what it does:

- `np.fft.fft` computes a 1D discrete Fourier transform
- `np.fft.fft2` computes a 2D discrete Fourier transform
- `np.fft.ifft` and `np.fft.ifft2` go back to physical space

So the FFT is the tool that numerically performs the change of basis.


## 6. Small 1D example: what FFT output means

We first look at a tiny 1D example.

If a function is made from just a few sine and cosine waves, then only a few Fourier coefficients should be large.


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
n_demo = 128
x_demo = np.linspace(0.0, 2.0 * np.pi, n_demo, endpoint=False)

f_demo = np.sin(3 * x_demo) + 0.5 * np.cos(5 * x_demo)
fhat_demo = np.fft.fft(f_demo)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x_demo, f_demo)
plt.title('Function sampled on a grid')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.stem(np.abs(fhat_demo), basefmt=' ')
plt.title('Magnitude of FFT coefficients')
plt.xlabel('mode index')
plt.ylabel('|f_hat|')
plt.tight_layout()
plt.show()
```

In the FFT plot, only a few modes are large. That means the function is mostly built from only a few Fourier basis functions.


## 7. The 2D heat equation in Fourier space

If

$$u(x,y,t)=\sum_{k,\ell} \hat u_{k,\ell}(t)e^{i(kx+\ell y)},$$

then substituting into

$$u_t = \kappa (u_{xx}+u_{yy})$$

gives

$$\frac{d\hat u_{k,\ell}}{dt} = -\kappa (k^2+\ell^2) \hat u_{k,\ell}.$$

So instead of one coupled PDE, we get many independent scalar ODEs.

Each one has exact solution

$$\hat u_{k,\ell}(t) = e^{-\kappa (k^2+\ell^2)t}\hat u_{k,\ell}(0).$$

That is the whole numerical method.


## 8. Set up the 2D periodic grid

We now move to the full 2D problem on $[0,2\pi)\times[0,2\pi)$.


```python
Lx = 2.0 * np.pi
Ly = 2.0 * np.pi
nx = 128
ny = 128

x = np.linspace(0.0, Lx, nx, endpoint=False)
y = np.linspace(0.0, Ly, ny, endpoint=False)
dx = x[1] - x[0]
dy = y[1] - y[0]

X, Y = np.meshgrid(x, y, indexing='xy')
kappa = 0.08

print('Grid:', nx, 'x', ny)
print('dx =', dx)
print('dy =', dy)
```

## 9. Choose a clear initial condition: a hot disk

We use a disk of temperature 1 in the center and 0 outside.

This is a good example because it is easy to understand in physical space, and its sharp boundary creates many Fourier modes.


```python
def initial_condition_circle(X, Y, cx=np.pi, cy=np.pi, radius=0.9):
    R = np.sqrt((X - cx)**2 + (Y - cy)**2)
    u0 = np.zeros_like(X)
    u0[R <= radius] = 1.0
    return u0


u0 = initial_condition_circle(X, Y)
```

```python
plt.figure(figsize=(5.5, 4.5))
plt.imshow(u0, origin='lower', extent=[0, Lx, 0, Ly])
plt.title('Initial temperature: hot disk')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
plt.tight_layout()
plt.show()
```

## 10. Change basis numerically: compute the 2D FFT

Now we move from physical space to Fourier space.

- `u0` is the temperature field in physical space
- `u0_hat = fft2(u0)` is the same field represented in the Fourier basis

So `u0_hat[k,l]` is the coefficient of one Fourier mode.


```python
u0_hat = np.fft.fft2(u0)
print('Shape of u0_hat:', u0_hat.shape)
```

## 11. Visualize the disk in Fourier space

The Fourier coefficients are complex numbers, so we usually plot their magnitude.

We use `fftshift` so that low frequencies appear in the center of the image.

Interpretation:

- bright near the center means strong low-frequency content
- spread away from the center means higher-frequency content
- sharp edges in physical space tend to create broader Fourier spectra


```python
u0_hat_shifted = np.fft.fftshift(u0_hat)
spectrum0 = np.log10(np.abs(u0_hat_shifted) + 1e-12)

plt.figure(figsize=(5.5, 4.5))
plt.imshow(spectrum0, origin='lower')
plt.title('log10 magnitude of Fourier coefficients')
plt.xlabel('shifted k-index')
plt.ylabel('shifted l-index')
plt.colorbar()
plt.tight_layout()
plt.show()
```

## 12. Build the wave number arrays

The exact Fourier-space heat solution uses the factor

$$e^{-\kappa (k^2+\ell^2)t}.$$

So we need the wave numbers in each coordinate direction.


```python
kx = 2.0 * np.pi * np.fft.fftfreq(nx, d=dx)
ky = 2.0 * np.pi * np.fft.fftfreq(ny, d=dy)
KX, KY = np.meshgrid(kx, ky, indexing='xy')
K2 = KX**2 + KY**2

print('K2 shape:', K2.shape)
```

## 13. Solve the PDE in Fourier space

The mode-by-mode exact formula is

$$\hat u_{k,\ell}(t) = e^{-\kappa (k^2+\ell^2)t}\hat u_{k,\ell}(0).$$

Then we apply the inverse FFT to return to physical space.


```python
def heat_solution_fft_2d(u0, K2, kappa, t):
    """Return the 2D heat equation solution at time t using the FFT."""
    u0_hat = np.fft.fft2(u0)
    decay = np.exp(-kappa * K2 * t)
    u_hat_t = decay * u0_hat
    u_t = np.fft.ifft2(u_hat_t)
    return np.real(u_t)
```

## 14. Plot the solution at several times

This should show the disk spreading and the sharp boundary becoming smooth.


```python
times = [0.0, 0.05, 0.15, 0.4, 1.0, 3.0]

fig, axes = plt.subplots(2, 3, figsize=(11, 7))
axes = axes.ravel()

for ax, t in zip(axes, times):
    ut = heat_solution_fft_2d(u0, K2, kappa, t)
    ax.imshow(ut, origin='lower', extent=[0, Lx, 0, Ly], vmin=0.0, vmax=1.0)
    ax.set_title(f't = {t}')
    ax.set_xticks([])
    ax.set_yticks([])

fig.suptitle('2D heat diffusion from a hot disk')
plt.tight_layout()
plt.show()
```

## 15. Compare physical space and Fourier space at a later time

At a later time:

- the disk is smoother in physical space
- the higher-frequency Fourier content is smaller

This is the diffusion mechanism in both representations.


```python
t = 0.4
u_t = heat_solution_fft_2d(u0, K2, kappa, t)
u_hat_t = np.exp(-kappa * K2 * t) * u0_hat
spectrum_t = np.log10(np.abs(np.fft.fftshift(u_hat_t)) + 1e-12)

fig, axes = plt.subplots(1, 2, figsize=(10, 4.5))
axes[0].imshow(u_t, origin='lower', extent=[0, Lx, 0, Ly], vmin=0.0, vmax=1.0)
axes[0].set_title(f'Physical space at t={t}')
axes[0].set_xticks([])
axes[0].set_yticks([])

im = axes[1].imshow(spectrum_t, origin='lower')
axes[1].set_title(f'Fourier magnitude at t={t}')
axes[1].set_xticks([])
axes[1].set_yticks([])
plt.colorbar(im, ax=axes[1])

plt.tight_layout()
plt.show()
```

## 16. Why does diffusion smooth?

Every Fourier mode is multiplied by

$$e^{-\kappa (k^2+\ell^2)t}.$$

So:

- low-frequency modes decay slowly
- high-frequency modes decay quickly

For the hot disk, the sharp boundary contains many higher frequencies. Those decay first, so the boundary softens over time.


```python
t = 0.6
decay = np.exp(-kappa * K2 * t)

plt.figure(figsize=(5.5, 4.5))
plt.imshow(np.fft.fftshift(decay), origin='lower')
plt.title(f'Decay factor at t={t}')
plt.colorbar()
plt.tight_layout()
plt.show()
```

## 17. Animation

This cell creates an in-notebook animation of the heat diffusion process.


```python
t_final = 2.0
num_frames = 80
time_grid = np.linspace(0.0, t_final, num_frames)

fig, ax = plt.subplots(figsize=(5.5, 4.5))
im = ax.imshow(u0, origin='lower', extent=[0, Lx, 0, Ly], vmin=0.0, vmax=1.0)
title = ax.set_title('t = 0.0')
ax.set_xticks([])
ax.set_yticks([])

def update(frame):
    t = time_grid[frame]
    ut = heat_solution_fft_2d(u0, K2, kappa, t)
    im.set_data(ut)
    title.set_text(f't = {t:.3f}')
    return [im, title]

ani = FuncAnimation(fig, update, frames=num_frames, interval=70, blit=False)
plt.close(fig)
HTML(ani.to_jshtml())
```

## 18. Optional single-mode check

If the initial condition is exactly one Fourier mode,

$$u(x,y,0)=\sin(mx)\cos(\ell y),$$

then the exact solution is

$$u(x,y,t)=e^{-\kappa(m^2+\ell^2)t}\sin(mx)\cos(\ell y).$$

This gives a clean correctness check.


```python
m = 3
ell = 4
t = 0.9

u_mode = np.sin(m * X) * np.cos(ell * Y)
u_fft = heat_solution_fft_2d(u_mode, K2, kappa, t)
u_exact = np.exp(-kappa * (m**2 + ell**2) * t) * np.sin(m * X) * np.cos(ell * Y)

err = np.linalg.norm(u_fft - u_exact) / np.sqrt(nx * ny)
print('RMS error for single-mode test:', err)
```

## Final summary

This notebook used a **change of basis** to solve the 2D heat equation.

The flow was:

1. start with temperature values on a grid
2. change basis using the FFT
3. solve the PDE mode-by-mode in the Fourier basis
4. change back to physical space using the inverse FFT

Key ideas:

- the Fourier basis is natural for periodic problems
- the Laplacian becomes multiplication by $-(k^2+\ell^2)$
- the FFT is the algorithm that computes the basis change efficiently
- diffusion smooths by damping high-frequency modes
