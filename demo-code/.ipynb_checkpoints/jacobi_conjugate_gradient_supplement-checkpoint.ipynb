# Jacobi iteration and conjugate gradient

This notebook is a compact supplement on two iterative methods for solving

$$
Ax=b.
$$

The goal is not to list formulas. The goal is to explain the mechanism.

Jacobi is a stationary fixed-point iteration. It is simple, local, and cheap per step.

Conjugate gradient is a structured optimization method for symmetric positive definite systems. It uses the geometry induced by $A$ far more effectively.

The emphasis is on three pictures:

- a linear system as geometry,
- a quadratic energy whose minimizer solves the system,
- the role of conditioning in both convergence and failure.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Why iterative methods appear

A direct method tries to solve the system in a finite sequence of eliminations.

An iterative method starts from an initial guess $x^{(0)}$ and produces a sequence

$$
x^{(0)},x^{(1)},x^{(2)},\dots
$$

that ideally converges to the exact solution $x_\ast$.

This is natural when $A$ is large and sparse. In that setting, matrix-vector products are cheap, while factorization can be expensive or destroy sparsity.

## Jacobi as a fixed-point map

Split the matrix as

$$
A=D+(L+U),
$$

where $D$ is diagonal and $L+U$ contains the off-diagonal entries.

Then

$$
Dx=b-(L+U)x,
$$

which suggests the iteration

$$
x^{(k+1)} = D^{-1}\bigl(b-(L+U)x^{(k)}\bigr).
$$

This is a fixed-point method

$$
x^{(k+1)} = T_J x^{(k)} + c,
\qquad
T_J=-D^{-1}(L+U).
$$

The method converges for every initial guess if the powers of $T_J$ decay to zero. A standard condition is

$$
\rho(T_J)<1.
$$

So Jacobi is controlled by how strong the off-diagonal coupling is relative to the diagonal.

## The geometric picture

For a $2\times 2$ system, the two equations are two lines.

The solution is their intersection.

Jacobi updates one component at a time using the old value of the other component. In two dimensions this typically creates a zig-zag path toward the intersection. When the coupling is weak, the zig-zag contracts. When the coupling is too strong, it can stall or diverge.

```python
A = np.array([[4., 1.],
              [2., 3.]])
b = np.array([1., 2.])

def jacobi_path(A, b, x0, steps):
    D = np.diag(np.diag(A))
    R = A - D
    x = x0.astype(float).copy()
    xs = [x.copy()]
    for _ in range(steps):
        x = np.linalg.solve(D, b - R @ x)
        xs.append(x.copy())
    return np.array(xs)

xs = jacobi_path(A, b, np.array([0., 0.]), 8)
xstar = np.linalg.solve(A, b)

x1 = np.linspace(-0.2, 0.6, 200)
y1 = (b[0] - A[0,0]*x1)/A[0,1]
y2 = (b[1] - A[1,0]*x1)/A[1,1]

plt.figure(figsize=(5,5))
plt.plot(x1, y1, label='row 1')
plt.plot(x1, y2, label='row 2')
plt.plot(xs[:,0], xs[:,1], 'o-')
plt.plot(xstar[0], xstar[1], 'x', ms=10)
plt.axis('equal')
plt.grid()
plt.legend()
plt.show()
```

The path is generated by repeated action of the iteration matrix on the error.

If $e^{(k)}=x^{(k)}-x_\ast$, then

$$
e^{(k+1)} = T_J e^{(k)}.
$$

So Jacobi is fundamentally an error propagation process.

## A structural success condition

A common sufficient condition is strict diagonal dominance:

$$
|a_{ii}| > \sum_{j\neq i} |a_{ij}|
\qquad \text{for each } i.
$$

This condition is not necessary, but it is informative.

It says each unknown is primarily controlled by its own equation. The off-diagonal terms act as corrections, not as the dominant mechanism. Jacobi performs well in exactly that regime because it treats the diagonal part as the easy piece and the off-diagonal part as a lagged perturbation.

## A failure mode

Jacobi can fail when the off-diagonal coupling is too strong.

Then the fixed-point map is not contractive. The method keeps reusing stale information from the previous iterate, and the correction can overshoot instead of reducing the error.

That is not a coding issue. It is a structural failure of the iteration matrix.

```python
def jacobi_residuals(A, b, steps):
    D = np.diag(np.diag(A))
    R = A - D
    x = np.zeros_like(b, dtype=float)
    vals = []
    for _ in range(steps):
        vals.append(np.linalg.norm(b - A @ x))
        x = np.linalg.solve(D, b - R @ x)
    return np.array(vals)

A_good = np.array([[4., 1.],
                   [2., 3.]])
A_bad = np.array([[1., 2.],
                  [2., 1.]])
b = np.array([1., 1.])

rg = jacobi_residuals(A_good, b, 12)
rb = jacobi_residuals(A_bad, b, 12)

plt.figure(figsize=(5,3))
plt.semilogy(rg, 'o-', label='convergent case')
plt.semilogy(rb, 'o-', label='failure case')
plt.grid()
plt.legend()
plt.show()
```

## Jacobi on a model tridiagonal system

Consider the tridiagonal matrix

$$
A =
\begin{bmatrix}
2 & -1 \\
-1 & 2 & -1 \\
& \ddots & \ddots & \ddots \\
&& -1 & 2
\end{bmatrix}.
$$

This is symmetric positive definite. It is the standard one-dimensional finite-difference Laplacian.

Jacobi does converge here, but slowly.

The reason is instructive: Jacobi only uses diagonal information in the solve step, so it damps some error components much faster than others. High-frequency error is reduced relatively well. Smooth error is reduced slowly. This is one reason Jacobi is useful as a smoother in multigrid, but not usually as a final solver.

```python
n = 40
A = 2*np.eye(n) - np.eye(n,1) - np.eye(n,-1)
b = np.ones(n)

rj = jacobi_residuals(A, b, 40)

plt.figure(figsize=(5,3))
plt.semilogy(rj, 'o-')
plt.grid()
plt.show()
```

## From linear algebra to optimization

Now assume $A$ is symmetric positive definite.

Define the quadratic functional

$$
\phi(x)=\frac12 x^T A x - b^T x.
$$

Then

$$
\nabla \phi(x)=Ax-b.
$$

So the equation $Ax=b$ is equivalent to the first-order optimality condition

$$
\nabla \phi(x)=0.
$$

That means solving the linear system is the same as minimizing a strictly convex quadratic.

## Why the quadratic picture matters

In two dimensions, the level sets of $\phi$ are ellipses.

If $A$ has widely separated eigenvalues, those ellipses are very stretched. That is the geometric meaning of poor conditioning.

A steepest descent step moves in the direction of $- \nabla \phi(x)$, which is the residual direction $r=b-Ax$. On a stretched ellipse, steepest descent often zig-zags because each step corrects strongly in one metric but not in the geometry that really matters.

```python
A = np.array([[3., 1.],
              [1., 2.]])
b = np.array([1., 1.])

def steepest_descent(A, b, x0, steps):
    x = x0.astype(float).copy()
    xs = [x.copy()]
    for _ in range(steps):
        r = b - A @ x
        a = (r @ r)/(r @ A @ r)
        x = x + a*r
        xs.append(x.copy())
    return np.array(xs)

def cg_path(A, b, x0, steps):
    x = x0.astype(float).copy()
    r = b - A @ x
    p = r.copy()
    xs = [x.copy()]
    for _ in range(steps):
        a = (r @ r)/(p @ A @ p)
        x = x + a*p
        xs.append(x.copy())
        r_new = r - a*(A @ p)
        beta = (r_new @ r_new)/(r @ r)
        p = r_new + beta*p
        r = r_new
    return np.array(xs)

sd = steepest_descent(A, b, np.array([-0.8, 1.2]), 6)
cg = cg_path(A, b, np.array([-0.8, 1.2]), 2)
xstar = np.linalg.solve(A, b)

xx = np.linspace(-1.2, 1.2, 300)
yy = np.linspace(-0.2, 1.6, 300)
X, Y = np.meshgrid(xx, yy)
Z = 0.5*(A[0,0]*X**2 + 2*A[0,1]*X*Y + A[1,1]*Y**2) - b[0]*X - b[1]*Y

plt.figure(figsize=(5,5))
plt.contour(X, Y, Z, levels=12)
plt.plot(sd[:,0], sd[:,1], 'o-', label='steepest descent')
plt.plot(cg[:,0], cg[:,1], 'o-', label='CG')
plt.plot(xstar[0], xstar[1], 'x', ms=10)
plt.axis('equal')
plt.grid()
plt.legend()
plt.show()
```

The plot shows the mechanism.

Steepest descent follows the Euclidean gradient direction. It decreases the energy, but it can waste steps by repeatedly correcting directions that interact strongly.

Conjugate gradient uses search directions that are $A$-orthogonal:

$$
p_i^T A p_j = 0
\qquad \text{for } i\neq j.
$$

So each step adds a new independent correction in the geometry induced by $A$.

## The CG recurrences

Starting from $x_0$, set

$$
r_0=b-Ax_0,
\qquad
p_0=r_0.
$$

Then for $k=0,1,2,\dots$,

$$
\alpha_k = \frac{r_k^T r_k}{p_k^T A p_k},
$$

$$
x_{k+1}=x_k+\alpha_k p_k,
$$

$$
r_{k+1}=r_k-\alpha_k A p_k,
$$

$$
\beta_k = \frac{r_{k+1}^T r_{k+1}}{r_k^T r_k},
$$

$$
p_{k+1}=r_{k+1}+\beta_k p_k.
$$

These formulas are short, but they encode two strong facts:

$$
r_i^T r_j = 0 \quad (i\neq j),
$$

and

$$
p_i^T A p_j = 0 \quad (i\neq j).
$$

Residuals are mutually orthogonal, and search directions are mutually conjugate.

## A more technical interpretation

At step $k$, CG searches in the affine Krylov space

$$
x_0 + \mathcal K_k(A,r_0),
\qquad
\mathcal K_k(A,r_0)=\operatorname{span}\{r_0,Ar_0,\dots,A^{k-1}r_0\}.
$$

Among all points in that space, CG picks the one that minimizes the $A$-norm of the error:

$$
\|e\|_A = \sqrt{e^T A e}.
$$

So CG is not just a descent method. It is an optimal projection process over an expanding sequence of Krylov subspaces.

## Conditioning enters again

For symmetric positive definite $A$, the condition number is

$$
\kappa(A)=\frac{\lambda_{\max}(A)}{\lambda_{\min}(A)}.
$$

This measures the eccentricity of the quadratic level sets.

A standard estimate is

$$
\|e_k\|_A \le
2\left(
\frac{\sqrt{\kappa(A)}-1}{\sqrt{\kappa(A)}+1}
\right)^k
\|e_0\|_A.
$$

So CG is fast when the spectrum is clustered and slower when the eigenvalues are widely spread.

```python
def cg_residuals(A, b, steps):
    x = np.zeros_like(b)
    r = b - A @ x
    p = r.copy()
    vals = [np.linalg.norm(r)]
    for _ in range(steps):
        a = (r @ r)/(p @ A @ p)
        x = x + a*p
        r_new = r - a*(A @ p)
        vals.append(np.linalg.norm(r_new))
        beta = (r_new @ r_new)/(r @ r)
        p = r_new + beta*p
        r = r_new
    return np.array(vals)

n = 40
A = 2*np.eye(n) - np.eye(n,1) - np.eye(n,-1)
b = np.ones(n)

rj = jacobi_residuals(A, b, 40)
rcg = cg_residuals(A, b, 40)

plt.figure(figsize=(5,3))
plt.semilogy(rj/rj[0], 'o-', label='Jacobi')
plt.semilogy(rcg/rcg[0], 'o-', label='CG')
plt.grid()
plt.legend()
plt.show()
```

On this model problem, both methods are legal, but they exploit very different amounts of structure.

Jacobi uses only the diagonal inverse and a lagged correction.

CG uses matrix-vector products, residual orthogonality, conjugate directions, and the energy geometry of the SPD system.

That is why the convergence profiles are so different.

## Final summary

Jacobi iteration is a fixed-point method based on the splitting $A=D+(L+U)$.

Its behavior is governed by the iteration matrix

$$
T_J=-D^{-1}(L+U).
$$

It is easy to implement and cheap per step, but it can fail when coupling is too strong and it is typically slow on smooth error components.

Conjugate gradient applies to symmetric positive definite matrices.

It solves the equivalent quadratic minimization problem by building conjugate search directions in Krylov subspaces.

The method is still cheap per step for sparse matrices, but it uses the operator geometry much more effectively.

The deeper lesson is that iterative methods are not distinguished only by cost per iteration. They are distinguished by what structure of the operator they preserve and exploit.
